### Алгоритм по извлечению текста из картинки

````python
# Картинка из патента https://www.surechembl.org/patent/EP-2810938-A1
from pathlib import Path
from PIL import Image
import pytesseract

img_path = Path("data/img/molecules.jpg")
img = Image.open(img_path)

custom_cfg = r"--psm 4"
extracted_text = pytesseract.image_to_string(
    img,
    lang="eng",
    config="--psm 4",
)

print(extracted_text.strip())
````
